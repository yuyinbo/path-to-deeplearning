# 创建张量

```python
# 1.使用列表创建Tensor
t1 = torch.Tensor([1, 2, 3])
print(t1)
# tensor([1., 2., 3.])

# 2.使用numpy数组创建Tensor
array1 = np.arange(12).reshape(3, 4)
t2 = torch.Tensor(array1)
print(t2)
```

```python
t2[0:1,:] #切片(左闭右开, 先行后列)
```

```python
# 3.通过torch的API创建Tensor
# torch.empty(3,4)：创建3行四列的空的tensor,会用无用的数据进行填充
# torch.ones([3,4]):三行四列全为1的tensor
# torch.zeros([3,4]):三行四列全为0的tensor
# torch.rand([3,4]):三行四列随机值在[0,1]之间的值
# torch.randint(low = 0, high = 10, size = [3, 4]) 创建3*4的随机整数的Tensor，值区间：[low, high]
# torch.randn([3,4]) 均值为0，方差为1,3*4的tensor
print(torch.empty(3,4))
print(torch.ones([3,4]))
print(torch.zeros([3,4]))
print(torch.rand([3,4]))
```

# 张量的方法和属性

```python
# 1.tensor.item(),当tensor中只有一个元素
a =torch.tensor(np.arange(1))
print(a)
print(a.item()) #只有一个元素的时候可以用,返回一个python的标量,不可以用于多个元素的tensor
print('-'*50)
print(torch.Tensor([[[1]]]).item()) #获取一个tensor中的元素值
# torch.Tensor([[[1]]])是一个三维张量, 但是1*1*1的, 只有一个元素
print('-'*50)
```

```python
# 2.Tensor转为ndarray
t2 = torch.Tensor([[3,4]])
print(t2.numpy()) #tensor转为ndarray
print('-'*50)
print(t2.shape) #获取形状
print(t2.size()) #获取形状
print(t2.size(1))  # 获取指定维度的长度, 这里0轴长度是1, 1轴长度是2
```

```python
array1 = np.array([[1,2,3],[4,5,6]])
print(id(array1))
array2=array1.reshape(3,2)
print(id(array2))
```

```python
array2[0,0]=100
print(array1)
print(array2)
# 虽然reshape是创建了新对象, 但是只是浅copy
```

```python
t2 = torch.Tensor([[[3,4]]])
print(t2.size())
# 形状是1*1*2
print(t2.view([1,2])) 
# [1,2] 表示1行2列, 变成2维
print(t2.view([2])) 
# 变成1维
```

```python
b=t2.view([2, -1]) 
# -1表示自动计算
print(b)
print('-'*50)
print(t2) 
# t2的形状并没有发生改变
print(f"id(t2)={id(t2)}")
print(f"id(b)={id(b)}")
# 两者id不同
t2.untyped_storage().untyped().data_ptr() == b.untyped_storage().untyped().data_ptr() 
# 判断两个tensor是否共享内存
# 但是两者共享内存空间
```

```python
b[0,0]=100
print(b)
print(t2) 
# tensor的view也是浅拷贝
```

```python
#3. 获取维数
print(t2.dim())
#4.获取最大值
print(t2.max())
# 返回的是一个0维张量
#5.转置
t3 = torch.tensor([[1,3,4], [2,4,6]])
print(t3)
print(t3.t())
```

```python
# 交换轴,这里的permute, rollaxis功能类似
t4 = torch.tensor(np.arange(24).reshape(2,3,4))
print(t4.shape)
print("-"*50)
print(t4.transpose(0, 1).shape)
# 交换0轴和1轴
print("-"*50)
print(t4.permute(1, 0, 2).shape)
# 将1轴滚动到0的位置
print("-"*50)
print(t4.permute(1, 2, 0).shape)
# 变为了3*4*2
print("-"*50)
print(t4.permute(2, 1, 0).shape)
# 变为了4*3*2
t4.type()
```

```python
# 在方法后加_，会原地修改
x = torch.tensor(np.arange(12).reshape(3,4),dtype = torch.int8)
print(x)
y= torch.ones([3,4], dtype = torch.int64)
print(y)

print('-'*50)
print(f"id(x)={id(x)}")
x.sub_(y) 
# x逐元素减去y
print(f"id(x)={id(x)}")
# 加上_就地修改，不加_的会创建一个新的tensor来存储
print(x)
```

```python
#tensor取值,和np完全相同
t5 = torch.tensor(np.arange(12).reshape(3,4))
print(t5)
print(t5[1,2]) 
# 取值
print(t5[1])
# 取一行
print(t5[:,1]) 
# 取一列
print(t5[1:3,1:3]) 
# 取一部分
print(t5[1:3,:]) 
# 取一部分
```

```python
#两个16行，1列的张量相减，求均值
t6 = torch.tensor(np.arange(16).reshape(16,1),dtype=torch.float32)
t7 = torch.tensor(np.arange(16,32).reshape(16,1),dtype=torch.float32)
print(t6)
print(t7)
(t6-t7).mean()
```

```python
#初始化两个张量，一个3,4，一个3,1,运算与ndarray相同(广播)
t8 = torch.tensor(np.arange(12).reshape(3,4),dtype=torch.float32)
t9 = torch.tensor(np.arange(3).reshape(3,1),dtype=torch.float32)
print(t8)
print(t9)
t8-t9
```

```python
# GPU中tensor的使用：
# 1.实例化device， torch.device("cpu" or "cuda:0"),选cuda:0需要torch.cuda.is_available()==true
# 2.tensor.to(device),把tensor转化为CUDA支持的tensor或者CPU支持的tensor
x = torch.tensor(np.arange(12).reshape(3,4),dtype = torch.int8)

print(torch.cuda.is_available())
if torch.cuda.is_available():
    device = torch.device("cuda") 
    # cuda device对象
    y = torch.ones_like(x, device = device) 
    # 创建一个在cuda的tensor
    x = x.to(device) 
    # 把x转为cuda上的tensor
    z = x+y
    print(z)
    print(z.to("cpu", torch.double))
    # 把z转为cpu上的tensor, 转换为double

x=torch.tensor([1.9806], device = "cuda:0")
device = torch.device("cpu")
x.to(device)
# 将GPU里的Tensor转到CPU
```

```python
t8 = torch.tensor(np.arange(12).reshape(3,4),dtype=torch.float32)
t9 = torch.tensor(np.arange(12).reshape(3,4),dtype=torch.float32)
print(t8)
print(t9)
t8*t9
# 直接相乘是逐元素相乘
```

```python
torch.mm(t8, t9.t())
# mm, Matrix Multiplication
# 三行四列并不能直接乘三行四列, 需要转置
```
