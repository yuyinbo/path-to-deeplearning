# 分类估计器

做数据预处理的是转换器, 做分类/回归的是估计器

使用KNN(k-近邻)预测用户签到位置

优点:  
1.训练时间复杂度为0  
2.既可以做分类也可以做回归, 算法简单,理论成熟  
3.可用于非线性分类  
缺点:  
预测的时间复杂度和空间复杂度高, 对稀有类别的预测准确度低

```python
data = pd.read_csv("./data/FBlocation/train.csv")

print(data.head(10))
print(data.shape)
data.info()
```

```python
# 只处理部分数据
data = data.query("x > 1.0 &  x < 1.25 & y > 2.5 & y < 2.75")
# query用字符串写类似SQL的条件表达式
# 等价于
#  data = data[
#     (data["x"] > 1.0) & (data["x"] < 1.25) &
#     (data["y"] > 2.5) & (data["y"] < 2.75)
# ]
# 使用 & 而不是and, 因为是逐元素运算
print(data.shape)
print(data.describe())
data
# 筛选了之后行号不是连续的了
```

```python
time_value = pd.to_datetime(data['time'], unit='s')
# 将时间戳转换为时间
print(time_value.head(10))  
```

```python
# 把日期格式转换成 索引格式
time_value = pd.DatetimeIndex(time_value)
print(time_value[0:10])
```

```python
# 时间本身作为特征不合适, 我们更加关心签到时间在一个月内是几号, 在一个星期内是星期几, 在一天内是几点
# 虽然脱敏过, 但是分布不会改变
print(type(data))

data.insert(data.shape[1], 'day', time_value.day) 
data.insert(data.shape[1], 'hour', time_value.hour)
data.insert(data.shape[1], 'weekday', time_value.weekday) 
# insert相当于直接赋值, data.shape[1]表示插入到最后
# data['day'] = time_value.day
# data['hour'] = time_value.hour
# data['weekday'] = time_value.weekday
data = data.drop(['time'], axis=1)
# 删除时间戳
print('-' * 50)
data.head()
```

```python
data.describe()
```

```python
place_count = data.groupby('place_id').count()
print(place_count)
# 有805个不同的签到位置, 有些地点打卡数非常多
```

```python
place_count['x'].describe()
# 有50%的签到位置签到数小于两次
```

```python
tf = place_count[place_count.x > 5].reset_index()
# 筛选出签到数大于5的地点
tf 
# 只剩下188个签到地点
```

```python
data = data[data['place_id'].isin(tf.place_id)]
data
# 筛选过后仍然剩下16694个样本, 即188个签到地点包含了大部分样本
```

```python
# 取出目标值
y = data['place_id']
# 一般设目标值为y
x = data.drop(['place_id'], axis=1)
x = x.drop(['row_id'], axis=1)
# 删除无用的特征值(row_id)和目标值
print(x.shape)
print(x.columns)
```

预处理完成

```python
# 在给出的完整数据中, 分割训练集和测试集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)

std = StandardScaler()
x_train = std.fit_transform(x_train)
# 对训练集特征值进行标准化

print(std.mean_)
print(std.var_)
x_test = std.transform(x_test)  
print(id(x_test))
# transform 不再进行均值和方差的计算，是在原有的基础上去标准化
# 返回的是副本
print('-' * 50)
print(id(x_test))
print(std.mean_)
print(std.var_)
# 使用的仍是fit得到的参数
```

```python
x_train.shape
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
# KNN分类器, 参数n_neighbors是超参数, 纳入多少个邻居
knn.fit(x_train, y_train)
# knn的fit是不训练的，只是把训练集的特征值和目标值放入到内存中
y_predict = knn.predict(x_test)
# 计算发生在预测阶段, 每个点都找最近三个点投票
# 得出预测结果
print("预测的签到位置：", y_predict[0:10])
print("预测的准确率:", knn.score(x_test, y_test))
# 得出准确率, score指用x_test预测出的y_predict去和y_test比较
# 等价于
# from sklearn.metrics import accuracy_score
# accuracy_score(y_test, knn.predict(x_test))
```

```python
print(y_predict[0:10])
print(y_test.to_numpy()[0:10])
```

# 网格搜索-调整超参数

```python
param = {"n_neighbors": [3, 5, 10, 12, 15],'weights':['uniform', 'distance']}
# weights参数是距离权重，uniform是距离权重相同，distance是距离越近权重越大
gc = GridSearchCV(knn, param_grid=param, cv=3)
# 将多个参数组成5*2个组合, 进行网格搜索，cv=3是3折交叉验证，用其中2折训练，1折验证
gc.fit(x_train, y_train)
# 给出的训练集, 又被分为训练集, 验证集

print("在测试集上最好的准确率：", gc.score(x_test, y_test))
# 在测试集上最好的准确率
print("在交叉验证当中最好的准确率：", gc.best_score_)
# 三折交叉验证的最好的准确率
print("选择最好的模型是：", gc.best_estimator_) 
print("每个超参数每次交叉验证的结果：")
gc.cv_results_
# 训练集分成三折, 以每折为验证集的情况下, 有10个参数组合, 共30个结果
```

# 朴素贝叶斯文本分类

优点:
1.来源于古典数学理论,分类效率稳定  
2.对缺失数据不太敏感, 常用于文本分类(指的是缺少段落而不是NaN)  
3.分类速度快,准确率高  
缺点: 先验概率无法准确得知

```python
news = fetch_20newsgroups(subset='all', data_home='data')

print(len(news.data))  
print('-'*50)
print(news.data[0]) 
print('-'*50)
print(news.target[0:5]) 
print(np.unique(news.target)) 
# 标签一共有哪些类别
print(news.target_names) 
# 标签的名字
```

```python
x_train, x_test, y_train, y_test = train_test_split(news.data, news.target, test_size=0.25, random_state=1)
# 分割训练集和测试集
tf = TfidfVectorizer()
print(len(x_train))
x_train = tf.fit_transform(x_train)
# 得到训练集每个词的TF-IDF值, 值越大表示越重要
print(len(tf.get_feature_names_out()))
# 一共出现了多少个特征(单词)
```

```python
x_train.shape
```

```python
print(tf.get_feature_names_out()[100000])
```

```python
import time
mlt = MultinomialNB(alpha=1.0)
# 进行朴素贝叶斯算法的预测, 公式是：P(y|x) = P(y) * P(x|y)/P(x)
# y指一个标签, x指一组特征, 给出一组特征, 是某个标签的概率, 等于该标签的先验概率, 乘以该标签下特征出现的概率, 除以特征出现的概率
# 因为特征是给定的, 所以忽略P(x)
# 训练是计算每种标签-特征组合的概率, 得到出现每个特征时, 是每个类别的概率
# alpha是拉普拉斯平滑系数，计算每个特征的 P(x|y)时, 分子加上alpha，分母加alpha*特征词数目
# 防止某类中没出现过某个特征, 遇到该特征时算出概率0
# 以防预测集中某篇文章, 无论多么符合某一类, 但因训练集中该类不包含该文章中的某个词, 完全无法匹配到该类
# print(x_train.toarray()[0])
# 训练
start=time.time()
mlt.fit(x_train, y_train) 
# 训练模型, 传入的是一个稀疏存储的矩阵, 自动转换成了array
end=time.time()
end-start #统计训练时间
```

```python
x_transform_test = tf.transform(x_test)  
# 使用训练集的特征处理测试集, 遇到新词会被丢弃
print(len(tf.get_feature_names_out()))
```

```python
x_transform_test.shape
```

```python
start=time.time()
y_predict = mlt.predict(x_transform_test)
print("前十篇文章预测的类别为：", y_predict[0:10])
print("前十篇文章真实的类别为：", y_test[0:10])
print("准确率为：", mlt.score(x_transform_test, y_test))
end=time.time()
end-start 
#预测时间
```

```python
print(classification_report(y_test, y_predict,
      target_names=news.target_names))
# 计算对于每个标签的精确率, 召回率, F1分数, support指事实上是每个标签的样本数
# 精确率是预测为真的结果中，实际为真的概率, TP/(TP+FP)
# 召回率是实际为真的结果中，预测为真的概率, TP/(TP+FN)
# F1分数是2TP/(2TP+FP+FN)
```

```python
# 对类别0演示精确率, 召回率和F1分数的算法
y_test1 = np.where(y_test == 0, 1, 0)
# 在测试集中, 如果标签是0, 则改为1, 否则改为0
print(y_test1.sum())
# 得到的是测试集属于类别0的样本数(TP+FN)
y_predict1 = np.where(y_predict == 0, 1, 0)
print(y_predict1.sum())
# 得到预测集属于类别0的样本数(TP+FP)
```

```python
t=(y_test1*y_predict1).sum()
print(t)
# 只有两个都为1, 才是TP
```

```python
print('精确率:', t/(y_predict1.sum()))
print('召回率:', t/(y_test1.sum()))
print('F1分数:', 2*t/(y_predict1.sum()+y_test1.sum()))
```

```python
y_test1 = np.where(y_test == 0, 1, 0)
print(y_test1.sum()) 
y_predict1 = np.where(y_predict == 0, 1, 0)
print(y_predict1.sum())
# roc_auc_score的y_test只能是二分类,针对多分类如何计算AUC
print("AUC指标：", roc_auc_score(y_test1, y_predict1))
```

# 决策树

ID3(香农熵)计算信息增益, 会偏向特征值多的特征, 使特征值多的特征更可能放在根节点  
CD4.5 计算信息增益率, 减小对多值特征的偏向, 但是计算复杂   
CART使用基尼指数  
基尼值公式是1-sum(p_i^2), 即根据某项特征分类后, 信息的"纯度"(越小越纯)  
基尼系数就是基尼值乘以概率权重    
对每个特征尝试分裂, 如果分裂前后基尼指数下降最多(纯度提升最多), 则用该特征分裂

优点:  
容易解释和可视化  
不需要很多的数据预处理, 不需要归一化和标准化   
缺点:  
不能使用过于复杂的树, 容易过拟合   
很小的数据变化就会导致完全不同的树(弱分类器)

对决策树进行参数调优

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
dict1 = DictVectorizer(sparse=False)
# 注意不需要归一化, 因为算的是概率

x_train = dict1.fit_transform(x_train.to_dict(orient="records"))
x_test = dict1.transform(x_test.to_dict(orient="records"))

# print(x_train)

dec = DecisionTreeClassifier(max_depth=7,min_impurity_decrease=0.01,min_samples_split=20)
# 调整决策树超参数, max_depth最大深度, 越大越容易过拟合
# min_impurity_decrease纯度改善, 只有纯度改善大于0.01才会分裂, 越小越容易过拟合
# min_samples_split最小样本数, 一个节点至少有20个样本才会尝试分裂, 越小越容易过拟合
# 调整这三个参数, 实际上是"预剪枝"
dec.fit(x_train, y_train)
print("预测的准确率：", dec.score(x_test, y_test))

export_graphviz(dec, out_file="tree1.dot",
                feature_names=['age', 'pclass=1st', 'pclass=2nd',
                'pclass=3rd', 'female', 'male'])
```

# 随机森林

适用于特征复杂, 样本多的情况(泰坦尼克太小了)

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
dict2 = DictVectorizer(sparse=False)

x_train = dict2.fit_transform(x_train.to_dict(orient="records"))
x_test = dict2.transform(x_test.to_dict(orient="records"))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=-1)
# n_jobs=-1, 使用所有CPU
param = {
    "n_estimators": [100, 300, 500],
    # 树的数量
    "max_depth": [3, 5, 7, 10],
    # 最大深度
    "min_samples_split": [5, 10, 20],
    # 分裂的最小样本数
    "min_samples_leaf": [1, 2, 5],
    # 叶子节点的最小样本数
    "max_features": ["sqrt", "log2", None]
    # 随机选取特征的个数, 总特征数的平方, 总特征数的对数, 取全部特征
    # 但在特征只有三个的情况下, 一般取全部特征, 实际上退化成普通树
    # 还可以使用的参数:
    # bootstrap: 是否有放回抽样(默认为True), 否则使用全部数据
    # 注意是有放回, 大约只会选中63.2%的数据
    # max_samples: 每棵树使用多少样本(配合bootstrap=True)
}
# 网格搜索(选参数)和交叉验证(算结果)
gc = GridSearchCV(rf, param_grid=param, cv=3)

gc.fit(x_train, y_train)

print("最好的准确率：", gc.score(x_test, y_test))

print("选择的参数：", gc.best_params_)
```
