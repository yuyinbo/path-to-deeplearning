# 处理缺失数据

```python
df_data = pd.DataFrame([np.random.randn(3), [1., 2., np.nan],
                       [np.nan, 4., np.nan], [1., 2., 3.]])
print(df_data.head())
```

```python
print(df_data.isnull())
# 返回一个内容为bool型的DataFrame
print(df_data.isnull().sum()/df_data.shape[0])
# 每一列的缺失率
# 每一列的bool类型求和, 除以行数
```

删除缺失数据

```python
print(df_data.dropna(subset=[0], inplace=False))
#subset=[0]是指按第一列来删除,第一列有空值就删除对应的行
#inplace=False，不改变原DataFrame
```

```python
print(df_data.dropna(axis=1))
# 按照轴1来删除, 即如果某一列有NaN, 就删除该列
```

填充缺失数据

```python
print(df_data.iloc[:,0].fillna(-100.))
# 将0列的缺失值填充为-100.0, 返回的是一列
print(df_data.fillna(-100.))
# 将每个缺失值都填充为-100.0
```

# 分组运算

```python
dict_obj = {'key1' : ['a', 'b', 'a', 'b',
                      'a', 'b', 'a', 'a'],
            'key2' : ['one', 'one', 'two', 'three',
                      'two', 'two', 'one', 'three'],
            'data1': np.random.randint(1, 10, 8),
            'data2': np.random.randint(1, 10, 8)}
df_obj = pd.DataFrame(dict_obj)
print(df_obj)
```

```python
# 按key1分组后，计算data1，data2的统计信息并附加到原始表格中，并添加表头前缀
k1_sum = df_obj.groupby('key1').mean(numeric_only=True).add_prefix('mean_')
# numeric_only=True, 只对只包含数值的列进行运算
# 添加前缀mean_
print(k1_sum)
```

```python
k1_sum_tf= df_obj.groupby('key1').transform('mean', numeric_only=True).add_prefix('mean_')
# transform指对每个列进行变换,最终形状与原DataFrame相同
# 作用是将每组对应的行的值替换为该组平均值
print(k1_sum_tf)
```

```python
def diff_mean(s):
    """
    返回数据与均值的差值
    """
    return s - s.mean()

print(df_obj.loc[:, ['key1','data1', 'data2']].groupby('key1').transform(diff_mean))
# 选定每一行里的key1列和data1, data2列, 对key1列进行分组, 对每一组每个元素取diff_mean
```

# 处理重复值

```python
df_obj = pd.DataFrame({'data1' : ['a'] * 4 + ['b'] * 4,
                       'data2' : np.random.randint(0, 4, 8)})
print(df_obj)
```

```python
print(df_obj.duplicated())
# 返回一个布尔Series, 长度与DataFrame相同
# 若一行的值是True, 说明和之前的某行重复了
```

```python
df_obj[~df_obj.duplicated()]
# 取出不重复的行(使用布尔索引)
```

```python
print(df_obj.duplicated('data2'))
# 按照某一列检查重复值
```
