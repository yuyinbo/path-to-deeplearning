# 轴

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)
# 最外层括号的是0轴, 第二层括号是1轴...
print('-' * 20)
print(np.sum(a, axis=0))
# 沿着轴0相加, 消掉第0维(把0轴对应位置的元素聚合起来)
print(np.sum(a, axis=1))
# 沿着轴1相加, 消掉第1维
print(np.sum(a))
```

```python
a = np.arange(60).reshape((3, 4, 5))
print(a)
b = np.sum(a, axis=0)
print("-" * 20)
print(b)
# 0轴的3块聚合起来, 变成4*5的
c = np.sum(a, axis=1)
print("-" * 20)
print(c)
d = np.sum(a, axis=2)
print("-" * 20)
print(d)
```

# 数值修改

```python
t = np.arange(24).reshape(4, 6)
t[1, :] = 0
# 修改第一行
t[:, 1] = 0
# 修改第一列
print(t)
t = np.arange(24).reshape(4, 6)
t[1:3, :] = 0
t[:, 1:3] = 0
# 修改多行多列
print(t)
t = np.arange(24).reshape(4, 6)
t[[0, 1], [2, 3]] = 0
# 将第0行第2列, 第1行第3列变为0(注意并不是0行1列)
print(t)
```

```python
t = np.arange(24).reshape(4, 6)
t[t > 10] = 0  # 将所有大于10的元素变为0
print(t)
t = np.arange(24).reshape(4, 6)
t[(t > 2) & (t < 6)] = 0  # 将大于2且小于6的
print(t)
t = np.arange(24).reshape(4, 6)
t[(t < 2) | (t > 6)] = 0  # 小于2或大于6的
print(t)
t = np.arange(24).reshape(4, 6)
t[~(t > 6)] = 0  # 不大于6的
print(t)
```

```python
t = np.arange(24).reshape(4, 6)
t = t.clip(10, 18)
# 将小于10的元素变为10,大于18的元素变为18, 防止元素过小或过大
print(t)
```

# 添加删除和去重

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)
print('\n')
print(np.append(a, [7, 8, 9]))
# 不指定轴的话, append会展平双方, 然后按列表添加
print('\n')
print(np.append(a, [[7, 8, 9]], axis=0))
# 沿轴0添加元素, 轴0会增长
print('\n')
print(np.append(a, [[5, 5, 5, 5], [7, 8, 9, 10]], axis=1))
# 轴1会增长
```

```python
a = np.array([[1, 2], [3, 4], [5, 6]])
print(a)
print('\n')
# 如果不指定轴会展开
print(np.insert(a, 3, [11, 12]))
print('\n')
# 指定了轴则会广播方式插入
print(np.insert(a, 1, [11], axis=0))
# 在轴0坐标为1的地方, 插入一个长度为2, 元素都为11的数组
print('\n')
print(np.insert(a, 1, 12, axis=1))
# 在轴1坐标为1的地方, 插入一个长度为3, 元素都为12的数组
```

```python
a = np.arange(12).reshape(3, 4)
print(a)
print('\n')
# 如果不指定轴会展开
print(np.delete(a, 5))
print('\n')
# 删除每一行中的1列：
print(np.delete(a, 1, axis=1))
print('\n')
```

统计数组中NaN的个数

```python
t = np.arange(24,dtype=float).reshape(4,6)
print(t)
t[3,4] = np.nan
# 将三行四列的数改成 nan
print(np.count_nonzero(t))
# count_nonzero() 函数返回非零元素的个数
print(np.count_nonzero(t != t))
# 先对每个元素运行t != t 生成掩码, 然后再统计0的个数
# 当任意一个元素不等于自己, 就知道是nan
t[np.isnan(t)] = 0
# 将 nan 替换为 0
print(t)
```

处理NaN

```python
t = np.arange(24).reshape(4,6).astype('float')
t[1,3:] = np.nan
# 将1行的3列后面全换成NaN
print(t)
print('-------------------')
for i in range(t.shape[1]):
    # 遍历每一列
    temp_col = t[:,i]
    nan_num = np.count_nonzero(temp_col != temp_col)
    # 判断该列是否有NaN, 统计有多少个
    if nan_num != 0:
        temp_col_not_nan = temp_col[temp_col == temp_col]
        # 将该列不为 nan 的数据赋给 temp_col_not_nan
        temp_col[np.isnan(temp_col)] = np.mean(temp_col_not_nan)
        # 将 nan 替换成这一列的平均值
print(t)
```
