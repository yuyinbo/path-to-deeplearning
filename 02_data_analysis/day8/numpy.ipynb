# 调整数组形状

```python
list4 = np.array([[1,2,3],[4,5,6]])
list4.shape = (3,2)
# 将2行3列变成3行2列
print(list4)
```

```python
list4 = np.array([[1,2,3],[4,5,6]])
print(f'id of list4:{id(list4)}')
list4.shape = (3,2)
# 修改了原来的数组
print(f'id of list4:{id(list4)}')
print(list4)
list5 = list4.reshape(3,2)
# 返回一个新的数组
print(f'id of list5:{id(list5)}')
print(list5)
```

```python
list6 = list4.reshape((6,),order='C')
# 将数组展开
# C表示按行优先，F表示按列优先,分别代表C语言风格和Fortran风格
print(list6)
```

```python
list7 = list4.flatten(order='C')
# 另一种展开的方法
print(list7)
```

# 轴

```python
a = np.array([[1,2,3],[4,5,6]])
print(a)
# 最外层括号的是0轴, 第二层括号是1轴...
print('-'*20)
print(np.sum(a,axis=0)) 
# 沿着轴0相加, 消掉第0维(把0轴对应位置的元素聚合起来)
print(np.sum(a,axis=1)) 
# 沿着轴1相加, 消掉第1维
print(np.sum(a))
```

```python
a=np.arange(60).reshape((3,4,5))
print(a)
b=np.sum(a, axis=0)
print("-"*20)
print(b)
# 0轴的3块聚合起来, 变成4*5的
c=np.sum(a, axis=1)
print("-"*20)
print(c)
d=np.sum(a, axis=2)
print("-"*20)
print(d)
```

# 索引和切片

```python
a = np.arange(10)
print(a[2:4])
print(a[2:])
print(a[0:10:2])
# 和list操作基本一样,左闭右开
```

```python
t1 = np.arange(24).reshape(4,6)
print(t1)
print('*'*20)
print(t1[1]) 
# 取一行
print('*'*20)
print(t1[1:])
# 取连续的多行
print('*'*20)
print(t1[1:3,:])
# 取连续的多行
print('*'*20)
print(t1[[0,2,3]])
# 取不连续的多行
print('*'*20)
print(t1[[0,2,3],:])
# 取不连续的多行
print('*'*20)
print(t1[:,1])
# 取一列
print('*'*20)
print(t1[:,1:])
# 连续的多列
print('*'*20)
print(t1[:,[0,2,3]])
# 取不连续的多列
print('*'*20)
print(t1[2,3])
# 取某一个值,三行四列
print('*'*20)
print(t1[2:4,3:5])
# 取一块区域
print('*'*20)
print(t1[[1,2], [0,2]])
# 取1行0列和2行2列
```

# 数值修改

```python
t = np.arange(24).reshape(4,6)
t[1,:]=0
# 修改第一行
t[:,1]=0
# 修改第一列
print(t)
t = np.arange(24).reshape(4,6)
t[1:3,:]=0
t[:,1:3]=0
# 修改多行多列
print(t)
t = np.arange(24).reshape(4,6)
t[[0,1],[2,3]]=0
# 将第0行第2列, 第1行第3列变为0(注意并不是0行1列)
print(t)
```

```python
t = np.arange(24).reshape(4,6)
t[t>10]=0  # 将所有大于10的元素变为0
print(t)
t = np.arange(24).reshape(4,6)
t[(t > 2) & (t < 6)] = 0  # 将大于2且小于6的
print(t)
t = np.arange(24).reshape(4,6)
t[(t < 2) | (t > 6)] = 0  # 小于2或大于6的
print(t)
t = np.arange(24).reshape(4,6)
t[~(t > 6)] = 0  # 不大于6的
print(t)
```

```python
t=np.arange(24).reshape(4,6)
t=t.clip(10,18)
# 将小于10的元素变为10,大于18的元素变为18, 防止元素过小或过大
print(t)
```

# 添加删除和去重

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)
print('\n')
print(np.append(a, [7, 8, 9]))
# 不指定轴的话, append会展平双方, 然后按列表添加
print('\n')
print(np.append(a, [[7, 8, 9]], axis=0))
# 沿轴0添加元素, 轴0会增长
print('\n')
print(np.append(a, [[5, 5, 5, 5], [7, 8, 9, 10]], axis=1))
# 轴1会增长
```

```python
a = np.array([[1, 2], [3, 4], [5, 6]])
print(a)
print('\n')
# 如果不指定轴会展开
print(np.insert(a, 3, [11, 12]))
print('\n')
# 指定了轴则会广播方式插入
print(np.insert(a, 1, [11], axis=0))
# 在轴0坐标为1的地方, 插入一个长度为2, 元素都为11的数组
print('\n')
print(np.insert(a, 1, 12, axis=1))
# 在轴1坐标为1的地方, 插入一个长度为3, 元素都为12的数组
```

```python
a = np.arange(12).reshape(3,4)
print(a)
print('\n')
# 如果不指定轴会展开
print(np.delete(a,5))
print('\n')
# 删除每一行中的1列：
print(np.delete(a,1,axis = 1))
print('\n')
```

去重

```python
a = np.array([5,2,6,2,7,5,6,8,2,9])
print (a)
print ('\n')
u = np.unique(a)
# 去重后的数组
print (u)
print ('\n')
```

```python
u,indices = np.unique(a, return_index = True)
# 每个元素在原数组里的索引
print (indices)
print ('\n')
print (u)
u,indices = np.unique(a,return_inverse = True)
# 返回原数组在去重数组中的下标
print (u)
print (indices)
print ('\n')
u,indices = np.unique(a,return_counts = True)
# 返回去重数组元素的重复次数
print (u)
print (indices)
```

数学计算

```python
score = np.array([[80,88],[82,81],[75,81]])

result = np.max(score)
# 获取所有数据最大值
print(result)

result = np.max(score,axis=0)
# 获取某一个轴上的数据最大值, 轴0消失, 只留下轴0每列最大值
print(result)

result = np.min(score)
# 获取最小值
print(result)
print()
result = np.min(score,axis=0)
# 获取某一个轴上的数据最小值, 轴0消失, 只留下轴0每列最小值
print(result)
# 数据的比较
result = np.maximum([-2, -1, 0, 1, 2], 0) 
# 第一个参数中的每一个数与第二个参数比较, 返回大的
print(result)
result = np.minimum([-2, -1, 0, 1, 2], 0) 
# 第一个参数中的每一个数与第二个参数比较, 返回小的
print(result)
result = np.maximum([-2, -1, 0, 1, 2], [1, 2, 3, 4, 5])
# 返回每个对应位置上的较大值
print(result)
print()
result = np.mean(score) # 获取所有数据的平均值
print(result)
result = np.mean(score,axis=0) # 获取某一行或者某一列的平均值
print(result)
print()
arr = np.array([[1,2,3], [4,5,6]])
print(arr)
print(arr.cumsum(0))
# 沿着轴0求前缀和, 即0行不变,1行变成与上一行的累加
print(arr.cumsum(1))
# 沿着轴1求前缀和, 即0行不变,1行变成与上一列的累加, 2行变成与之前列的累加
```

```python
print(score)
result = np.argmin(score,axis=0)
# 0轴最小值的索引
res=np.min(score,axis=0)
# 0轴最小值
print(result, res)
print()
result = np.std(score,axis=0)
# 0轴标准差
print(result)
result = np.ptp(score,axis=None)
# 所有元素的极差
print(result)
```
